```python
import pandas as pd
import numpy as np
```

```python
airline = pd.read_csv('Airlines.csv')
```

```python
airline
```

```python
airline.describe() # There are 923 Airlines but only 918 unique airlines. 923 tags and 923 unique tags. So, there are airlines with the same name but the tags being different
```

```python
# Remove the column that is not required
airline = airline.drop('unnamed',1)
airline
```

```python
airline['IATA'] = airline.Tag.str.split('/').str.get(0)
airline['ICAO'] = airline.Tag.str.split('/').str.get(1) 
```

```python
airline
```

```python
checknan = airline['ICAO'].isnull()
checknan
```

```python
i=0
for i in range(checknan.size):
    if (checknan[i]):
        airline['ICAO'][i] = airline['IATA'][i]
        airline['IATA'][i] = np.nan
```

```python
airline
```

```python
airline['Number'] = airline['Number'].str.rstrip('aircraft')
airline
```

```python
airline.to_csv('Airlines_callsign_IATA_ICAO.csv')
```
